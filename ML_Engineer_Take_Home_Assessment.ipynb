# MODEL IMPLEMENTATION:

```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np

from torchvision.models import resnet18, ResNet18_Weights
```

```python
class Model(nn.Module):

  def __init__(self):
    super(Model, self).__init__()
    pretrained_model = resnet18(weights=ResNet18_Weights.DEFAULT)
    self.backbone = nn.Sequential(*list(pretrained_model.children())[:-2])
    ### Initialize the required Layers
    self.feature_len = 7 * 7 * 512
    self.have_object = nn.Linear(self.feature_len, 2)
    self.cat_or_dog = nn.Linear(self.feature_len, 2)
    self.specie = nn.Linear(self.feature_len, 9)
    self.bbox = nn.Linear(self.feature_len, 4)
    ### Initialize the required Layers
    self.flatten = nn.Flatten()
  def forward(self, input):
      out_backbone = self.backbone(input)
      ### Write Forward Calls for the Model
      out = self.flatten(out_backbone)
      have_object_out = self.have_object(out)
      cat_or_dog_out = self.cat_or_dog(out)
      specie_out = self.specie(out)
      bbox_out = F.relu(self.bbox(out))
      

      return {
          "bbox": bbox_out,
          "object": have_object_out,
          "cat_or_dog": cat_or_dog_out,
          "specie": specie_out
      }
```

```python
from bs4 import BeautifulSoup

def read_xml_file(path):
    with open(path, 'r') as f:
        data = f.read()
    bs_data = BeautifulSoup(data, 'xml')
    return {
        "cat_or_dog": bs_data.find("name").text,
        "xmin": int(bs_data.find("xmin").text),
        "ymin": int(bs_data.find("ymin").text),
        "xmax": int(bs_data.find("xmax").text),
        "ymax": int(bs_data.find("ymax").text),
        "specie": "_".join(path.split(os.sep)[-1].split("_")[:-1])
    }
```

```python
from PIL import Image

import torchvision.transforms as T


class CustomDataset():

  def __init__(self, dataset_path, images_list, train=False):

    
    self.preprocess = T.Compose([
              T.Resize(256),
              T.CenterCrop(224),
              T.ToTensor(),
              T.Normalize(
                  mean=[0.485, 0.456, 0.406],
                  std=[0.229, 0.224, 0.225]
              )
    ])

    image_folder_path = os.path.join(dataset_path, "images")
    label_folder_path = os.path.join(dataset_path, "labels")
    self.cat_or_dog_mapping = {"cat": 0, "dog" : 1}
    self.specie_mapping = {'beagle' : 0, 
                           'pomeranian': 1, 
                           'basset_hound': 2, 
                           'chihuahua' : 3, 
                           'Persian': 4, 
                           'american_bulldog' : 5, 
                           'Abyssinian': 6, 
                           'Birman': 7, 
                           'american_pit_bull_terrier' : 8}

    self.labels = []
    self.image_paths = []
    for path in os.listdir(image_folder_path):
        name = path.split(os.sep)[-1].split(".")[0]
        if name in images_list:
          xml_path = os.path.join(label_folder_path, name+".xml")
          if os.path.exists(xml_path):
            xml_data = read_xml_file(xml_path)
          else:
            xml_data = -1
          self.labels.append(xml_data)
          if os.path.exists(os.path.join(image_folder_path, name + ".jpg")):
            imagepath = os.path.join(image_folder_path, name + ".jpg")
          else:
            imagepath = os.path.join(image_folder_path, name + ".jpeg")
          self.image_paths.append(imagepath)

  def __len__(self):
    return len(self.image_paths)

  def __getitem__(self, index):
    imagepath = self.image_paths[index]
    image = Image.open(imagepath)
    width, height = image.size
    image = self.preprocess(image)
    xml_data = self.labels[index]
    if xml_data == -1:
      label = {
          "bbox": torch.tensor([0, 0, 0, 0], dtype = torch.long),
          "object": torch.tensor([0], dtype = torch.long),
          "cat_or_dog": torch.tensor([-100], dtype = torch.long),
          "specie": torch.tensor([-100], dtype = torch.long),
          "width": torch.tensor([width], dtype = torch.long),
          "height": torch.tensor([height], dtype = torch.long)
      }
    else:
      x_min_norm = xml_data['xmin']/width
      x_max_norm = xml_data['xmax']/width
      y_min_norm = xml_data['ymin']/height
      y_max_norm = xml_data['ymax']/height
      label = {
          "bbox": torch.tensor([x_min_norm, y_min_norm, x_max_norm, y_max_norm], dtype = torch.long),
          "object": torch.tensor([1], dtype = torch.long),
          "cat_or_dog": torch.tensor([self.cat_or_dog_mapping[xml_data['cat_or_dog']]], dtype = torch.long),
          "specie": torch.tensor([self.specie_mapping[xml_data['specie']]], dtype = torch.long),
          "width": torch.tensor([width], dtype = torch.long),
          "height": torch.tensor([height], dtype = torch.long)
      }
    return image, label
```

```python
from torch.utils.data import DataLoader
```

```python
# import our library
from torchmetrics.detection.mean_ap import MeanAveragePrecision
from sklearn.metrics import accuracy_score

def custom_bbox_loss(input, labels):
  loss_fn = nn.MSELoss()
  bbox_loss = 0.
  batch_size = 0.
  labels_object = labels["object"].view(-1).detach().cpu().numpy().tolist()
  for i, label in enumerate(labels_object):
    if int(label) == 1:
      bbox_loss += loss_fn(input[i, :].view(-1), labels["bbox"][i,:].view(-1))
      batch_size += 1.
  
  return bbox_loss/batch_size

def train(epochs, model_weights = None):

  # Initialize Model and Optimizer
  # Resuming training
  device = "cuda" if torch.cuda.is_available() else "cpu"
  if model_weights is not None:
    model = Model()
    state_dict = torch.load(model_weights, map_location = device)
    model.load_state_dict(state_dict['model_state_dict'])
    model = model.to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr = 1e-3)
    optimizer.load_state_dict(state_dict['optimizer_state_dict'])
  else:
    model = Model()
    model = model.to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr = 1e-3)
  # Initialize Loss Functions
  have_object_loss = nn.CrossEntropyLoss()
  specie_loss = nn.CrossEntropyLoss()
  cat_or_dog_loss = nn.CrossEntropyLoss()
  bbox_loss = custom_bbox_loss # Not necessary you need to apply function to all coordinates together, You can have separete loss functions for all coordinates too

  training_dataset = CustomDataset("/content/assessment_dataset", images_list=train_list)
  training_loader = DataLoader(training_dataset, batch_size = 8, shuffle = True, drop_last = True)
  
  

  def train_one_epoch(epoch_index, tb_writer):
      device = "cuda" if torch.cuda.is_available() else "cpu"
      running_loss = 0.
      last_loss = 0.

      # Here, we use enumerate(training_loader) instead of
      # iter(training_loader) so that we can track the batch
      # index and do some intra-epoch reporting
      for i, data in enumerate(training_loader):
          # Every data instance is an input + label pair
          inputs, labels = data

          inputs = inputs.to(device).type(torch.float32)

          for k, v in labels.items():
            labels[k] = v.to(device).type(torch.float32)          

          # Make predictions for this batch
          outputs = model(inputs)

          # Compute the loss and its gradients
          loss_have_object = have_object_loss(outputs["object"], labels["object"].view(-1).type(torch.long))
          loss_specie = specie_loss(outputs["specie"], labels["specie"].view(-1).type(torch.long))
          loss_cat_or_dog = cat_or_dog_loss(outputs["cat_or_dog"], labels["cat_or_dog"].view(-1).type(torch.long))
          
          loss_bbox = bbox_loss(outputs["bbox"], labels)

          loss =  loss_have_object + loss_specie + loss_cat_or_dog + loss_bbox# Consolidate all individual losses
          loss.backward()
          optimizer.step()
          optimizer.zero_grad()

          # Gather data and report
          running_loss += loss.item()
          if i % 10 == 0:
              last_loss = running_loss / (i + 1) # loss per batch
              print(f"Loss : {last_loss}")
      return running_loss / len(training_loader)
  max_score_bbox = 0
  for i in range(epochs):

    epoch_loss = train_one_epoch(i, None)
    print(f' Epoch {i} Loss : {epoch_loss}')

    score_object, score_cat_or_dog, score_specie, score_bbox = test(model, val_list)
    print(f"Object Accuracy : {score_object}")
    print(f"Cat or dog accuracy: {score_cat_or_dog}")
    print(f"specie accuracy: {score_specie}")
    print(f"MAP score: {score_bbox}")
    if score_bbox > max_score_bbox:
      torch.save({
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
      }, "model.pth")
      max_score_bbox = score_bbox

def test(model, val_list):
  device = "cuda" if torch.cuda.is_available() else "cpu"

  def post_process_object(x, labels):
    object_preds = torch.argmax(x["object"], dim = -1).view(-1).cpu().numpy()
    object_true = labels["object"].view(-1).cpu().numpy()
    return object_preds, object_true

  def post_process_cat_or_dog(x, labels):
    cat_or_dog_preds = torch.argmax(x["cat_or_dog"], dim = -1).view(-1).cpu().numpy().tolist()
    cat_or_dog_true = labels["cat_or_dog"].view(-1).cpu().numpy().tolist()
    labels_obj = labels["object"].view(-1).cpu().numpy().tolist()
    preds = []
    y_true = []
    for pred, label_obj, true_label in zip(cat_or_dog_preds, labels_obj, cat_or_dog_true):
      if int(label_obj) == 1:
        preds.append(pred)
        y_true.append(true_label)
    return np.array(preds), np.array(y_true)

  def post_process_specie(x, labels):
    specie_preds = torch.argmax(x["specie"], dim = -1).view(-1).cpu().numpy().tolist()
    specie_true = labels["specie"].view(-1).cpu().numpy().tolist()
    labels_obj = labels["object"].view(-1).cpu().numpy().tolist()
    preds = []
    y_true = []
    for pred, label_obj, true_label in zip(specie_preds, labels_obj, specie_true):
      if int(label_obj) == 1:
        preds.append(pred)
        y_true.append(true_label)
    return np.array(preds), np.array(y_true)

  def post_process_bbox(x, labels):
    # Not evaluating per class - Everything is 0
    labels_obj = labels["object"].view(-1).cpu().numpy().tolist()
    preds, targets = [], []
    for i, label in enumerate(labels_obj):
      if int(label) == 1:
        # torchmetrics expects coordiantes in absolute format
        width = labels["width"][i].item()
        height = labels["height"][i].item()

        xmin_pred = max(x["bbox"][i,0].item(), 0) * width
        ymin_pred = max(x["bbox"][i,1].item(), 0) * height
        xmax_pred = max(x["bbox"][i,2].item(), 0) * width
        ymax_pred = max(x["bbox"][i,3].item(), 0) * height

        xmin_true = labels["bbox"][i,0].item() * width
        ymin_true = labels["bbox"][i,1].item() * height
        xmax_true = labels["bbox"][i,2].item() * width
        ymax_true = labels["bbox"][i,3].item() * height

        pred_label = 0
        true_label = 0

        score = torch.softmax(x["object"], dim = -1)[i,1].item()

        preds.append(dict(boxes=torch.tensor([[xmin_pred, ymin_pred, xmax_pred, ymax_pred]]),
                      scores=torch.tensor([score]),
                      labels=torch.tensor([1.])))
        targets.append(dict(boxes=torch.tensor([[xmin_true, ymin_true, xmax_true, ymax_true]]),
                      labels=torch.tensor([1.])))
      
    return preds, targets

  val_dataset = CustomDataset("/content/assessment_dataset", images_list=val_list)
  val_loader = DataLoader(val_dataset, batch_size = 8, shuffle = False)

  metric_object = accuracy_score
  metric_cat_or_dog = accuracy_score
  metric_specie = accuracy_score
  metric_bbox = MeanAveragePrecision()

  y_true_object = []
  y_pred_object = []

  y_true_cat_or_dog = []
  y_pred_cat_or_dog = []

  y_true_specie = []
  y_pred_specie = []
  for i, data in enumerate(val_loader):
    inputs, labels = data

    inputs = inputs.to(device)

    for k, v in labels.items():
      labels[k] = v.to(device)  

    # Make predictions for this batch
    outputs = model(inputs)

    preds_bbox, targets_bbox = post_process_bbox(outputs, labels)
    metric_bbox.update(preds_bbox, targets_bbox)

    object_preds, object_true = post_process_object(outputs, labels)
    y_true_object.append(object_preds)
    y_pred_object.append(object_true)

    cat_or_dog_preds, cat_or_dog_true = post_process_cat_or_dog(outputs, labels)
    y_true_cat_or_dog.append(cat_or_dog_preds)
    y_pred_cat_or_dog.append(cat_or_dog_true)

    cat_or_dog_preds, cat_or_dog_true = post_process_cat_or_dog(outputs, labels)
    y_true_cat_or_dog.append(cat_or_dog_preds)
    y_pred_cat_or_dog.append(cat_or_dog_true)

    specie_preds, specie_true = post_process_specie(outputs, labels)
    y_true_specie.append(specie_preds)
    y_pred_specie.append(specie_true)

  
  y_true_object = np.concatenate(y_true_object)
  y_pred_object = np.concatenate(y_pred_object)

  y_true_cat_or_dog = np.concatenate(y_true_cat_or_dog)
  y_pred_cat_or_dog = np.concatenate(y_pred_cat_or_dog)

  y_true_specie = np.concatenate(y_true_specie)
  y_pred_specie = np.concatenate(y_pred_specie)
  score_object = metric_object(y_true_object, y_pred_object)
  score_cat_or_dog = metric_cat_or_dog(y_true_cat_or_dog, y_pred_cat_or_dog)
  score_specie = metric_specie(y_true_specie, y_pred_specie)
  score_bbox = metric_bbox.compute()["map"].item()

  return score_object, score_cat_or_dog, score_specie, score_bbox
```

```python
from PIL import ImageDraw

def visualize(model_weights, image_folder_path = "/content/assessment_dataset/images", output_folder="output"):
  if not os.path.exists(output_folder):
    os.mkdir(output_folder)

  cat_or_dog_mapping = {"cat": 0, "dog" : 1}
  specie_mapping = {'beagle' : 0, 
                    'pomeranian': 1, 
                    'basset_hound': 2, 
                    'chihuahua' : 3, 
                    'Persian': 4, 
                    'american_bulldog' : 5, 
                    'Abyssinian': 6, 
                    'Birman': 7, 
                    'american_pit_bull_terrier' : 8}

  inv_mapping_cat_or_dog = {v:k for k, v in cat_or_dog_mapping.items()}
  inv_mapping_specie = {v:k for k, v in specie_mapping.items()}

  model_state_dict = torch.load(model_weights)
  model = Model()
  model.load_state_dict(model_state_dict['model_state_dict'])
  model.eval()

  device = "cuda" if torch.cuda.is_available() else "cpu"
  model = model.to(device)

  preprocess = T.Compose([
              T.Resize(256),
              T.CenterCrop(224),
              T.ToTensor(),
              T.Normalize(
                  mean=[0.485, 0.456, 0.406],
                  std=[0.229, 0.224, 0.225]
              )
  ])

  image_names = os.listdir(image_folder_path)
  final_return_dict = dict()
  with torch.no_grad():
    for image_name in image_names:
      image = Image.open(os.path.join(image_folder_path, image_name))

      width, height = image.size
      image_out = preprocess(image)
      image_out = image_out.unsqueeze(0)
      image_out = image_out.to(device)
      output = model(image_out)
      have_object = torch.argmax(output["object"], dim = -1).item()
      if int(have_object) == 1:
        has_object = True
        cat_or_dog = inv_mapping_cat_or_dog[torch.argmax(output["cat_or_dog"], dim = -1).item()]
        specie = inv_mapping_specie[torch.argmax(output["specie"], dim = -1).item()]

        bbox_out = output["bbox"]
        xmin = int(min(bbox_out[0, 0].item(), 1) * width)
        ymin = int(min(bbox_out[0, 1].item(), 1) * height)
        xmax = int(min(bbox_out[0, 2].item(), 1) * width)
        ymax = int(min(bbox_out[0, 3].item(), 1) * height)
        return_dict = { "has_object": has_object, "cat_or_dog": cat_or_dog, "specie": specie, "xmin": xmin, "ymin": ymin, "xmax": xmax, "ymax": ymax}
        draw = ImageDraw.Draw(image)
        draw.rectangle(((xmin, ymin), (xmax, ymax)), outline = "green", width = 3)
      else:
        return_dict = { "has_object": False, "cat_or_dog": "NA", "specie": "NA", "xmin": 0, "ymin": 0, "xmax": 0, "ymax": 0 }
      final_return_dict[image_name] = return_dict
      outfile = os.path.join(output_folder, image_name)
      image.save(outfile, "JPEG")
  return final_return_dict
```

```python
visualize(model_weights= "model.pth")
```
